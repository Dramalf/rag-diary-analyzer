```python
import time
import jwt
import requests

# 实际KEY，过期时间
def generate_token(apikey: str, exp_seconds: int):
    try:
        id, secret = apikey.split(".")
    except Exception as e:
        raise Exception("invalid apikey", e)

    payload = {
        "api_key": id,
        "exp": int(round(time.time() * 1000)) + exp_seconds * 1000,
        "timestamp": int(round(time.time() * 1000)),
    }
    return jwt.encode(
        payload,
        secret,
        algorithm="HS256",
        headers={"alg": "HS256", "sign_type": "SIGN"},
    )

url = "https://open.bigmodel.cn/api/paas/v4/chat/completions"
headers = {
  'Content-Type': 'application/json',
  'Authorization': generate_token("your_token", 1000)
}

data = {
    "model": "glm-3-turbo",
    "messages": [{"role": "user", "content": """你好"""}]
}

response = requests.post(url, headers=headers, json=data)

print("Status Code", response.status_code)
print("JSON Response ", response.json())
```

```python
# 打开并读取文本文件
filename = 'artical.txt'
content=[]
# 使用with语句确保文件在使用完毕后会被自动关闭
with open(filename, 'r', encoding='utf-8') as file:
    # 逐行读取文件内容
    for line in file:
        line=line.strip()
        if line and line!="图片" and line!="关注该公众号" and "\u200b" not in line:
            content.append(line)

print(content)
```

```python
articals=[]
idx=0
articals.append({
    "title":"",
    "meta":"",
    "lines":[]
})

for line in content:
    if line =="微信扫一扫":
        idx+=1
        articals.append({
            "title":"",
            "meta":"",
            "lines":[]
        })
        continue
    else:
        if articals[idx]["title"] =="":
            articals[idx]["title"]=line
        elif articals[idx]["meta"] =="":
            articals[idx]["meta"]=line
        else:
            articals[idx]["lines"].append(line)
articals
            
    
```

```python
for artical in articals:
    artical["content"]=''.join(artical["lines"])
```

```python
import jieba
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import normalize
```

```python
questions=[
    {
        "question":"为什么感到焦虑和迷茫",
        "answer": "",
        "reference": ""
    },
        {
        "question":"压力来源于哪里",
        "answer": "",
        "reference": ""
    },
       {
        "question":"客户的爱好是什么",
        "answer": "",
        "reference": ""
    }
]
```

```python


question_words = [' '.join(jieba.lcut(x['question'])) for x in questions]
pdf_content_words = [' '.join(jieba.lcut(''.join(x['lines']))) for x in articals]

tfidf = TfidfVectorizer()
tfidf.fit(question_words + pdf_content_words)

question_feat = tfidf.transform(question_words)
pdf_content_feat = tfidf.transform(pdf_content_words)

question_feat = normalize(question_feat)
pdf_content_feat = normalize(pdf_content_feat)

for query_idx, feat in enumerate(question_feat):
    score = feat @ pdf_content_feat.T
    score = score.toarray()[0]
    max_score_page_idx = score.argsort()[-1]
    questions[query_idx]['reference'] = 'artical_' + str(max_score_page_idx)
```

```python
def ask_glm(content):
    url = "https://open.bigmodel.cn/api/paas/v4/chat/completions"
    headers = {
      'Content-Type': 'application/json',
      'Authorization': generate_token("your_token", 1000)
    }

    data = {
        "model": "glm-3-turbo",
        "messages": [{"role": "user", "content": content}]
    }

    response = requests.post(url, headers=headers, json=data)
    return response.json()
```

```python
# !pip install rank_bm25
from rank_bm25 import BM25Okapi

pdf_content_words = [jieba.lcut(''.join(x['lines'])) for x in articals]
bm25 = BM25Okapi(pdf_content_words)

for query_idx in range(len(questions)):
    doc_scores = bm25.get_scores(jieba.lcut(questions[query_idx]["question"]))
    max_score_page_idx = doc_scores.argsort()[-1]
    questions[query_idx]['reference'] = 'page_' + str(max_score_page_idx)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer("./m3e-base")
```

```python
question_sentences = [x['question'] for x in questions]
pdf_content_sentences = [x['content'] for x in articals]

question_embeddings = model.encode(question_sentences, normalize_embeddings=True)
pdf_embeddings = model.encode(pdf_content_sentences, normalize_embeddings=True)


```

```python
pdf_embeddings
```

```python
for query_idx, feat in enumerate(question_embeddings):
    score = feat @ pdf_embeddings.T
    max_score_page_idx = score.argsort()[-1]
    questions[query_idx]['reference'] = 'artical_' + str(max_score_page_idx)
    prompt = '''你是一个心理专家助手，擅长通过客户的日记文本来分析客户的心理状态并给出相关建议。帮我结合给定的日记内容，回答一个问题。如果问题无法从资料中获得，请输出结合给定的资料，无法回答问题。
日记标题：{0}
日记内容：{1}

问题：{2}
    '''.format(
        articals[max_score_page_idx]['title'],
        articals[max_score_page_idx]['content'],
        questions[query_idx]["question"]
    )
    questions[query_idx]['aid']=max_score_page_idx;
    questions[query_idx]['prompt']=prompt
    print(prompt)
    # answer = ask_glm(prompt)#['choices'][0]['message']['content']
    # questions[query_idx]['answer'] = answer
```
